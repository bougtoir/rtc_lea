```python
import pandas as pd

# データフレームを読み込む
df = data

# データを文字列として扱うために変換
df['labor_date'] = df['labor_date'].astype(str)
df['labor_time'] = df['labor_time'].astype(str)
df['cervic_date'] = df['cervic_date'].astype(str)
df['cervic_time'] = df['cervic_time'].astype(str)
df['delivery_date'] = df['delivery_date'].astype(str)
df['delivery_time'] = df['delivery_time'].astype(str)

# 日時の変換でエラーが発生する行を特定して表示
error_rows = []

for index, row in df.iterrows():
    try:
        pd.to_datetime(row['labor_date'].split(' ')[0] + ' ' + row['labor_time'], format="%Y-%m-%d %H:%M:%S")
        pd.to_datetime(row['cervic_date'].split(' ')[0] + ' ' + row['cervic_time'], format="%Y-%m-%d %H:%M:%S")
        pd.to_datetime(row['delivery_date'].split(' ')[0] + ' ' + row['delivery_time'], format="%Y-%m-%d %H:%M:%S")
    except ValueError as e:
        error_rows.append(index)
        print(f"Error in row {index}: {e}")

# エラーが発生する行を削除
df_cleaned = df.drop(error_rows)

# 結果の確認
# print("Cleaned DataFrame:")
# print(df_cleaned)

# ここで、クリーンなデータフレームに対して、フェーズの計算を続けることができます。
df_cleaned['labor_datetime'] = pd.to_datetime(df_cleaned['labor_date'].str.split(' ').str[0] + ' ' + df_cleaned['labor_time'])
df_cleaned['cervic_datetime'] = pd.to_datetime(df_cleaned['cervic_date'].str.split(' ').str[0] + ' ' + df_cleaned['cervic_time'])
df_cleaned['delivery_datetime'] = pd.to_datetime(df_cleaned['delivery_date'].str.split(' ').str[0] + ' ' + df_cleaned['delivery_time'])

# phase1の計算（分単位）
df_cleaned['phase1'] = (df_cleaned['cervic_datetime'] - df_cleaned['labor_datetime']).dt.total_seconds() / 60

# phase2の計算（分単位）
df_cleaned['phase2'] = (df_cleaned['delivery_datetime'] - df_cleaned['cervic_datetime']).dt.total_seconds() / 60

# delivery_durationの計算（分単位）
df_cleaned['delivery_duration'] = (df_cleaned['delivery_datetime'] - df_cleaned['labor_datetime']).dt.total_seconds() / 60

# 結果の確認
print(df_cleaned[['phase1', 'phase2', 'delivery_duration']])
```
